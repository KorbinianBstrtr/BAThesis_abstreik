```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def visualize_train(train_result):
    plt.plot(train_result["epochs"], train_result["train_loss"], label='Training')
    plt.plot(train_result["epochs"], train_result["val_loss"], label='Validation')
    if len(train_result["test_loss"]):
        plt.plot(train_result["epochs"], train_result["test_loss"], label='Test')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend(loc='best')
    plt.show()
```

```python
df = pd.read_csv("train_limited_data_results.csv", index_col="n_points_train_val")
df["train_result"] = df.apply(lambda row: eval(row["train_result"]), axis=1)
```

```python
plt.plot(df.index, df.apply(lambda x: min(x["train_result"]["test_loss"]), axis=1))
plt.show()
```

```python
visualize_train(df["train_result"][5])
```
